```python
D[1,:]
np.corrcoef(D[1,:],C)
np.corrcoef(D[1,:],C)[0,1]
N, k = D.shape
rL = []
for i in range(N):
    rL.append(np.corrcoef(D[i,:],C)[0,1])
print(rL)
rL = sorted(enumerate(rL), key=lambda x: -x[1])
print(rL)
r = [x[1] for x in rL]
L = [x[0] for x in rL]
```

```python
N = len(L)
S_mask = np.zeros(N)
print(S_mask)
S_mask[gene_sets[0]] = 1
print(S_mask)
# reorder gene set mask
S_mask = S_mask[L]
print(S_mask)
p_exp = 1
N_R = sum(abs(r * S_mask)**p_exp)
print(N_R)
if N_R != 0:
    P_hit = np.cumsum(abs(r * S_mask)**p_exp) / N_R
    print(P_hit)
else:
    P_hit = np.zeros_like(S_mask)
N_H = len(gene_sets[0])
if N != N_H:
    P_mis = np.cumsum((1 - S_mask)) / (N - N_H)
    print(P_mis)
else:
    P_mis = np.zeros_like(S_mask)
idx = np.argmax(abs(P_hit - P_mis))
np.zeros_like(S_mask)
```
